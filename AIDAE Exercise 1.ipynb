```
def CheckPrime(x):
  try:
    if x<2:
      return False
    for i in range(2,x):
      if x%i ==0:
        return False
    return True    
  except Exception as e:
    print("my program broke:", e)     
    
```

```
CheckPrime("a")
```

```
def Fibonacci(x, doubleTheValues=False):
  series=[]
  for i in range(0,x):
    if i==0 or i==1:
      series.append(1)
    else:
      series.append(series[-2]+series[-1])

  return series
    
```

```
Fibonacci(3)
```
